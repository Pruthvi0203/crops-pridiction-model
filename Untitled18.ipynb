```python
import joblib

model = joblib.load("crop_predict_model.pkl")
scaler = joblib.load("scaler.pkl")
le = joblib.load("label_encoder.pkl")
```

```python
# Model details
print("Model Details:\n", model)
print("Number of Trees:", len(model.estimators_))
print("Feature Importance:", model.feature_importances_)
print("Crop Classes:", le.classes_)
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Input widgets
N_input = widgets.IntText(value=90, description='Nitrogen:')
P_input = widgets.IntText(value=42, description='Phosphorus:')
K_input = widgets.IntText(value=43, description='Potassium:')
Temp_input = widgets.FloatText(value=20, description='Temperature:')
Humidity_input = widgets.FloatText(value=80, description='Humidity:')
pH_input = widgets.FloatText(value=6.5, description='pH_Value:')
Rainfall_input = widgets.FloatText(value=200, description='Rainfall:')

button = widgets.Button(description="Predict Crop")
output = widgets.Output()

def predict_button_clicked(b):
    new_data = pd.DataFrame([[
        N_input.value,
        P_input.value,
        K_input.value,
        Temp_input.value,
        Humidity_input.value,
        pH_input.value,
        Rainfall_input.value
    ]], columns=['Nitrogen', 'Phosphorus', 'Potassium', 'Temperature', 'Humidity', 'pH_Value', 'Rainfall'])

    new_data_scaled = scaler.transform(new_data)
    predicted_crop = le.inverse_transform(model.predict(new_data_scaled))

    with output:
        output.clear_output()
        print("Predicted Crop:", predicted_crop[0])

button.on_click(predict_button_clicked)

display(N_input, P_input, K_input, Temp_input, Humidity_input, pH_input, Rainfall_input, button, output)
```
